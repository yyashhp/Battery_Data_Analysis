```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score

import warnings


warnings.filterwarnings('ignore')
data = pd.read_excel("WAM.xlsx")
```

```python
data.columns = ['dates', 'hour', 'price', 'load_forecast']
print(len(data))
data = data.dropna(axis = 0)
print(len(data))
```

```python
def season(row):
    month = int(str(data.iloc[1]['dates'])[5:7])
    day = int(str(data.iloc[1]['dates'])[8:10])
    if month>11 or month<3:
        return "Winter"
    if month == 11:
        if day > 16:
            return "Winter"
        else:
            return "Fall"
    if month == 3:
        if day > 16:
            return "Spring"
        else:
            return "Winter"
    if (month > 3 and month <6):
        return "Spring"
    if (month > 6 and month < 9):
        return "Summer"
    if month == 6:
        if day < 15:
            return "Spring"
        else:
            return "Summer"
    if month == 9:
        if day < 15:
            return "Summer"
        else:
            return "Fall"
    return "Fall"
def month(row):
    return int(str(row['dates'])[5:7])
def year(row):
    return int(str(row['dates'])[:4]) - 2017
```

```python
data['season'] = data.apply(season, axis = 1)
data['month'] = data.apply(month, axis  = 1)
data['is_winter'] = (data['season'] == "Winter")
data['is_fall'] = (data['season'] == "Fall")
data['is_spring'] = (data['season'] == "Spring")
data['is_summer'] = (data['season'] == "Summer")
data['year'] = data.apply(year, axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)
coded_season = one_hot_encoder.fit_transform(data["season"].values.reshape(-1, 1))
data['season_coded']=coded_season
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

predictors = ['hour', 'year','month', 'season_coded','load_forecast' ]

# Create a linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(train_data[predictors], train_data['price'])

# Make predictions on the test set
test_data['predict_price'] = model.predict(test_data[predictors])

# Evaluate the model
mse = mean_squared_error(test_data['price'], test_data['predict_price'])
print(f'Mean Squared Error: {mse}')

# Compare actual goal difference with predicted goal difference
test_data['PredictionError'] = test_data['price'] - test_data['predict_price']

data['predicted_price'] = model.predict(data[predictors])
print(model.coef_)
```

```python
from sklearn.metrics import r2_score
r2_score(test_data['price'], test_data['predict_price'])
r2_score(data['price'], data['predicted_price'])
```

```python
import math
import statistics

data['predicted_diff'] = data['predicted_price'].diff()
data.dropna(axis = 0, inplace = True)
data['running_diff'] = data['predicted_diff'].rolling(window = 24).mean()
print(len(data))
data.dropna(axis=0, inplace = True)

```

```python
from statistics import mean 


def trade(row):
    if row['predicted_diff'] < row['running_diff']*2:
        return "S"
    elif row['predicted_diff'] > row['running_diff']*2:
        return "B"
    else:
        return "N"
data['trade'] = data.apply(trade, axis = 1)

buys = []
sells = []
held = 0
buy_size = []
sell_size = []

for date, row in data.iterrows():
    if row['trade'] == 'N':
        continue
    elif row['trade'] == 'B':
        t = max(row['predicted_diff'],row['running_diff'])
        b = min(row['predicted_diff'],row['running_diff'])
        q = min(int(abs(t/b)), row['load_forecast'])
        q = int(q)
        #q = int(row['predicted_diff']/row['running_diff'])
        for num in range(q):
            buys.append(row['price'])
        held+=q
        buy_size.append(q)
    elif held>0 and row['trade'] == 'S':
        t = max(row['predicted_diff'],row['running_diff'])
        b = min(row['predicted_diff'],row['running_diff'])
        q = int(abs(t/b))**2
        #q = int(row['predicted_diff']/row['running_diff'])
        size = min(q, held, row['load_forecast'])
        size = int(size)
        for num in range(size):
            sells.append(row['price'])
        held-=size
        sell_size.append(size)
print(held)
if held>0:
    print(held)
    price = data.iloc[-1]['price']
    print(price)
    for num in range(held):
        sells.append(price)
    held = 0

profit = sum(sells)-sum(buys)
print(profit)
print(mean(sells))
print(mean(buys))
print(len(sells))
print(len(buys))
print(max(buy_size))
print(max(sell_size))

print(f'Profit: {profit}')
```
